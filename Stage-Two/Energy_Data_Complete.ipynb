### Task-4: Implementing the model
I intend to run aggregate analysis on my entire features using statsmodels.

```python
from pandas import DataFrame
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression
```

```python
feature_var = sm.add_constant(feature_var)
est = sm.OLS(target_var,feature_var).fit()

print(est.summary())
```

###### observations:
The table summary above, shows there are certain features that didnt fall within our level of confidence,which is a significant level of 0.05 as measured by the pvalue,ideally these fetured should be drop the model re-evaluated. The essence of running the statsmodel was to help us select the most important features that are significantly correlated with our target variable which is provided by the pvalue.

- rv1 and rv2 as we can see appear not have any correlation as they appear to be random values

- visibility and light columns will be dropped too despite meeting our condition,as they have so many outliers and add nothing to our analysis

- Kitchen_temp,RH_office, bathroom_temp will also be dropped as fail to meet our significant condition

- The feature variables were able to explain only 17% of variability on the target variable which is provided by the R2.

```python
feature_var.drop(['Visibility','lights'], axis = 1, inplace = True)
```

###### sepearating my data into train_test_split

```python
x_train,x_test,y_train,y_test = train_test_split(feature_var,target_var,test_size=0.3,random_state=42)
### fitting the model
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
predictions = model.predict(x_test)
```

###### I intend to calculate my model performance using the following models;

- Mean Avaerage Error(MAE)
- Residual Sum of Squares(RSS)
- Root Mean Square Error(RMSE)
- R-Squared Score(R2)

```python

from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, predictions)
print('MAE_SCORE PERFORMANCE : {:.2f}'.format(mae))
round(mae, 3)
print('=============================================')


from sklearn.metrics import r2_score

r2_score = r2_score(y_test, predictions)
print('R2_SCORE PERFORMANCE : {:.2f}'.format(r2_score))
round(r2_score, 3)
print('===============================================')

from sklearn.metrics import mean_squared_error

rmse = np.sqrt(mean_squared_error(y_test, predictions))
print('RMSE_SCORE PERFORMANCE : {:.2f}'.format(rmse))
round(rmse,3)
print('================================================')

import numpy as np
RSS = np.sum(np.square(y_test-predictions))
print('RSS_SCORE PERFORMANCE:')
round(RSS, 3)
```

### Task-5: Applying penalisation and Regularisation method
With all the models we have tried so far,they all seem to perform poorly.Now i intend to apply penalisation method by using the following algorithms to examine dataset features and reducing any form of overfitting.The following algorithms will be used

- Improved Linear Regression (*Ridge Regression*)

- Essemble Method (*Random Forest and Gradient Boosting Regression*)

- Neural Network Method (*MultiLayer Perceptron Regressor*)

```python
from sklearn.linear_model import Ridge
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor
from sklearn.linear_model import LinearRegression
from sklearn.neural_network import MLPRegressor
```

###### Measuring the accuracy the models

```python
 X_train,X_test,y_train,y_test=train_test_split(feature_df,target_df,test_size=0.3,random_state=1)

linear_reg = LinearRegression()
linear_reg.fit(X_train,y_train)
linear_preds = linear_reg.predict(X_test)

print('LinearReg_train_score: {:.2f}'.format(linear_reg.score(X_train,y_train)))
print('LinearReg_test_score: {:.2f}'.format(linear_reg.score(X_test,y_test)))
print('========================================')

ridge_reg = Ridge(alpha=0.5)
ridge_reg.fit(X_train,y_train)
ridge_preds = ridge_reg.predict(X_test)

print('Ridge_reg_train_score: {:.2f}'.format(ridge_reg.score(X_train,y_train)))
print('Ridge_Reg_test_score: {:.2f}'.format(ridge_reg.score(X_test,y_test)))
print('========================================')

forest_model = RandomForestRegressor(n_estimators=100)
forest_model.fit(X_train,y_train)
forest_preds = forest_model.predict(X_test)

print('RandomForest_train_score: {:.2f}'.format(forest_model.score(X_train,y_train)))
print('RandomForest_test_score: {:.2f}'.format(forest_model.score(X_test,y_test)))
print('========================================')

gradient_reg = GradientBoostingRegressor()
gradient_reg.fit(X_train,y_train)
gradient_preds = gradient_reg.predict(X_test)

print('Gradient_Reg_train_score: {:.2f}'.format(gradient_reg.score(X_train,y_train)))
print('Gradient_Reg_test_score: {:.2f}'.format(gradient_reg.score(X_test,y_test)))
print('========================================')

neural_reg = MLPRegressor()
neural_reg.fit(X_train,y_train)
neural_preds = neural_reg.predict(X_test)

print('Neural_Reg_train_score: {:.2f}'.format(neural_reg.score(X_train,y_train)))
print('Neural_Reg_test_score: {:.2f}'.format(neural_reg.score(X_test,y_test)))
```

###### plotting the residual graph for each of the above tested algorithms 

```python
f,ax = plt.subplots(2,2,figsize=(10,7))

vis1=sns.residplot(linear_preds,y_test, ax=ax[0][0])
vis2=sns.residplot(ridge_preds,y_test,ax=ax[0][1])
vis3=sns.residplot(forest_preds,y_test,ax=ax[1][0])
vis4=sns.residplot(gradient_preds,y_test,ax=ax[1][1])

vis1.title.set_text('ResidualPlots for Linear Regressor')
vis2.title.set_text('ResidualPlot for Ridge Regressor')
vis3.title.set_text('ResidualPlot for Random Forest')
vis4.title.set_text('ResidualPlot for Gradient Boosting')

plt.show()
```

```python
f,ax = plt.subplots(figsize=(5,3))
ax=sns.residplot(neural_preds,y_test)
ax.set_xlabel('Predicted')
ax.set_ylabel('Actual')

plt.title('ResidualPlot for Neural Regressor')
plt.show()
```

###### observations:
As we can see from the above residual plots that the worst performing model is the neural network regressor like we generated previously, followed by the ridge regressor and the linear regressor.The best performing model is clearly the random forest regressor, having almost a well balanced data between the predicted and the observed
